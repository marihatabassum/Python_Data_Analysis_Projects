```python
#Question 04: What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

more_than_50k_salary_df=adult_df[adult_df['salary']==' >50K']

masters= more_than_50k_salary_df[more_than_50k_salary_df['education']==' Masters']
#print('Percentage of people who has Bachelor degree make more than 50k and the % is: 28' )
bachelors=more_than_50k_salary_df[more_than_50k_salary_df['education']==' Bachelors']
doctorate=more_than_50k_salary_df[more_than_50k_salary_df['education']==' Doctorate']
advanced_edu= len(doctorate) +len(bachelors)+len(masters)
percentage_ae_50k=(advanced_edu*100)/len(more_than_50k_salary_df)

print("Percentage of people with advanced education who have more than 50k salary: ", round(percentage_ae_50k, 2), "%")
```

```python
#Quetion 05:  What percentage of people without advanced education make more than 50K?
without_ae=len(more_than_50k_salary_df)-advanced_edu
percentage_50k_without_ae=without_ae*100/len(more_than_50k_salary_df)
print("Percentage of people without advanced eduaction make more than 50k:" , round(percentage_50k_without_ae, 2), '%' )
```

```python
#Question 06: What is the minimum number of hours a person works per week?

adult_df['hours-per-week'].min()
```

```python
#Question 07: What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?


#more_than_50k_salary_df['hours-per-week'].value_counts()
percentage_min_hrs_50k=(len(more_than_50k_salary_df[more_than_50k_salary_df['hours-per-week']==1])*100)/len(more_than_50k_salary_df)
print("percentage of the people who work the minimum number of hours per week have a salary of more than 50K: ", round(percentage_min_hrs_50k, 3), '%')
```
